```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns 
```

```python
# Load the dataset
file_path = "Ship_Performance_Dataset.csv"
ship = pd.read_csv(file_path)
```

```python
#Shows first five rows
ship.head()
```

```python
#drops the cell with null values
ship.dropna()
```

```python
# Compute the correlation matrix
selected_columns = ["Engine_Power_kW", "Revenue_per_Voyage_USD", "Distance_Traveled_nm","Cargo_Weight_tons","Operational_Cost_USD","Speed_Over_Ground_knots"]
correlation_matrix = ship[selected_columns].corr()
```

```python
#To plot the HaetMap
plt.figure(figsize=(7, 6))
plt.title("Correlation Heatmap of Ship Performance Metrics", fontsize=14)
plt.xticks(rotation=45)  # Rotate column names for better readability
plt.yticks(rotation=0)
sns.heatmap(correlation_matrix,cmap="RdYlGn",annot=True)
plt.show()
```

```python
# Scatter Plots for Key Metrics
key_factors = ["Distance_Traveled_nm", "Cargo_Weight_tons", "Operational_Cost_USD"]
for factor in key_factors:
    plt.figure(figsize=(6, 4))
    sns.scatterplot(x=ship[factor], y=ship["Revenue_per_Voyage_USD"])
    plt.title(f"Revenue vs {factor}")
    plt.xlabel(factor)
    plt.ylabel("Revenue per Voyage (USD)")
    plt.show()
```
